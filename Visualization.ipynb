```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing = pd.read_csv("housing.csv")
```

```python
housing.head()
```

```python
sorted(housing.columns)
```

```python
sns.relplot(data = housing, x = "Gr Liv Area", y = "SalePrice")

plt.show()
```

```python
sns.set_theme()
sns.relplot(data = housing, x = "Gr Liv Area", y = "SalePrice",\
           hue = "Overall Qual")

plt.show()
```

```python
sns.set_theme()
sns.relplot(data = housing, x = "Gr Liv Area", y = "SalePrice",\
           hue = "Overall Qual", palette = "RdYlBu")

plt.show()
```

```python
sns.set_theme()
sns.relplot(data = housing, x = "Gr Liv Area", y = "SalePrice",\
           hue = "Overall Qual", palette = "RdYlBu", size = "Garage Area", sizes=(1,200), 
            style="Rooms",col = "Year")

plt.show()
```

```python
sns.set_theme()
sns.relplot(data = housing, x = "Gr Liv Area", y = "SalePrice",\
           hue = "Overall Qual", palette = "RdYlBu", size = "Garage Area", sizes=(1,200), 
            style="Rooms" , col="Rooms")

plt.show()
```
